```python
def log_plot(words,dataset_str):
    fdist = nltk.FreqDist(w.lower() for w in words)
    # print()
    most_common=fdist.most_common(1000)
    top1000values=[ y for x, y in most_common ]
    # print(ls)
    # print(most_common)
    # most_common.sort(reverse=True)
    rank = np.arange(1,len(most_common)+1)
    plt.figure()
    plt.plot(np.log(rank),np.log(top1000values))
    plt.xlabel("log(rank)")
    plt.ylabel("log(freq)")
    plt.title("dataset: "+dataset_str)
    plt.savefig(dataset_str+"_top1000.png")
    plt.show()
log_plot(tokens,"lec 1")
```

```python
## today we are not going to filter anything , everything will be considered 
def filter_words(dataset):
    words=dataset.words()
    stopls=stopwords.words('english')    
    filtered_words = [word.lower() for word in words if word.lower() not in stopls]
    return filtered_words
```

```python
def extract_ngrams(tokens, num):
    ret=[]
    for n in range(1,num):
        n_grams = ngrams(tokens, n)
        ls=[ ' '.join(grams) for grams in n_grams]
        ret.extend(ls)
    return ret
n_grams = extract_ngrams(tokens,10)
extract_ngrams(word_tokenize("hello world this is nirmal"),3)
```

```python
n_grams
```

```python
freq=Counter(n_grams)
```

```python
s="my experience with natural language processing"
def get_grams_starts_with_s(s,n_grams):
    ret=set()
    for gram in n_grams:
        if gram.startswith(s):
            ret.add(gram)
    return list(ret)
get_grams_starts_with_s("my",n_grams)
```

```python
sum(freq.values())
```

```python
# for 
# using chain rule for above s
# P(s) = P(my) * P(experience|my) * P(with|my experience) * P(natural|my experience with) * P(language|my experience with natural)*P(processing|my experience with natural language)
total_n_grams=sum(freq.values())
prob = freq.get("my")/total_n_grams * freq.get("experience")/freq.get("my") * freq.get("with")/freq.get("my experience") * freq.get("natural")/freq.get("my experience with")


print(prob)

    
```

```python
s="my"
while True:
    nextdt={}
    print("current string :", s)
    for next in get_grams_starts_with_s(s,n_grams):
        # next_gram=next.replace(s+" ","",1)
        nextdt[next]=freq.get(next)/freq.get(s)
    # print(nextdt)
    sorted_dict = dict(sorted(nextdt.items(), key=lambda item: item[1], reverse=True))
  
    for i in list(islice(sorted_dict.items(), 10)):
        print(i)
    print("enter your choice of word")
    w=input()
    s=s+" " +w
    # break
```

### POS tagging 

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

import os
transcribe=""
for path in os.listdir('lec_transcribe/'):
    f=open('lec_transcribe/'+path)
    lec_transcribe=f.read()
    transcribe=transcribe+"\n"+lec_transcribe
tokenized_dataset = [word_tokenize(sentence) for sentence in sent_tokenize(" ".join(transcribe.split("\n")))]

# Perform POS tagging
pos_tagged_dataset = [nltk.pos_tag(sentence) for sentence in tokenized_dataset]

# Print the POS tagged dataset
posdt={}
for tagged_sentence in pos_tagged_dataset:
    for i in range(0,len(tagged_sentence)-1):
        tr=(tagged_sentence[i][1],tagged_sentence[i+1][1])
        if tr in posdt.keys():
            posdt[tr]=posdt[tr]+1
        else:
            posdt[tr]=1
        # print(tagged_sentence[i][1],tagged_sentence[i+1][1])
        # break
    # break
    # print(tagged_sentence)
```

```python
for key, value in list(posdt.items()):
    if value<20:
        del posdt[key]
```
